```python
def top(df,n=5,column='tip_pct'):
    return df.sort_values(by=column)[-n:]
top(tips,n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker','day']).apply(top,n=3,column='total_bill')
```

```python
tips.groupby(['smoker','day'],group_keys=False).apply(top,n=3,column='total_bill')
```

```python
tips.groupby(['smoker','day'],as_index=False).apply(top,n=3,column='total_bill')
```

```python
frame=pd.DataFrame({'data1':np.random.randn(1000),
                    'data2':np.random.randn(1000)})
quartiles=pd.cut(frame.data1,4)
```

```python
quartiles[:10]
```

```python
def get_status(group):
    return {'min':group.min(),'max':group.max(),
            'count':group.count(),'mean':group.mean()}
grouped=frame.data2.groupby(quartiles)
grouped.apply(get_status).unstack()
```

```python
grouping = pd.qcut(frame.data1, 10)#, labels=False)
grouped = frame.data2.groupby(grouping)
grouped.apply(get_status)#.unstack()
```

```python
states = ['Ohio', 'New York', 'Vermont', 'Florida',
          'Oregon', 'Nevada', 'California', 'Idaho']
group_key=['west']*4+['east']*4#等价于 west west west west east east east east
data=pd.Series(np.random.randn(8),index=states)
data
```

```python
data['Ohio']=np.nan
fill_mean=lambda g:g.fillna(g.mean())

data.groupby(group_key).apply(fill_mean)
data.groupby(group_key).count()
```

```python
suits=['H','S','C','D']
card_val=(list(range(1, 11)) + [10] * 3) * 4
base_names=['A']+list(range(2,11))+['J','Q','K']
cards=[]
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names)
deck=pd.Series(card_val,index=cards)
def draw(deck,n=5):
    return deck.sample(n)
draw(deck)
```

```python
get_suit=lambda card:card[-1]
deck.groupby(get_suit).apply(draw,n=2)
```

```python
spx_corr=lambda x:x.corrwith(x['SPX'])
rets=close_px.pct_change().dropna()
get_year=lambda x:x.year
by_year=rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda g:g['AAPL'].corr(g['MSFT']))
```

```python
import statsmodels.api as sm
def regress(data,yvar,xvars):
    Y=data[yvar]
    X=data[xvars]
    X['intercept']=1.
    result=sm.OLS(Y,X).fit()
    return result.params
```

```python
by_year.apply(regress,'AAPL',['SPX'])
```
